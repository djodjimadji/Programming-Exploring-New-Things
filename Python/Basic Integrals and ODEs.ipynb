$\text{Exercise 1}$

Use python function <b> quad </b> to evaluate the following integrals

1. $A = \int_{-1}^1 4xdx$

```python
f=lambda x:4*x
A=quad(f,-1,1)[0]
A
```

2.  $B = \int_{-\pi}^{\pi} \cos (x)dx$

```python
g=lambda x:np.cos(x)
B=round(quad(g,-np.pi,np.pi)[0],0)
B
```

2.  $C = \int_{-\infty}^{\infty} \exp (-\dfrac{x^2}{2})dx$

```python
h=lambda x:np.exp(-x**2/2)
C=np.round(quad(h,-np.inf,np.inf)[0],2)
C
```

$\text{Exercise 2}$

Use python function <b> odeint </b> to solve and plot the solution of the initial value problem

\begin{equation}
  \begin{cases}
    y' = 2y + e^{2x},      & \quad \text{on } [0, 1]\\\\
    y(0) = 3,
  \end{cases}
\end{equation}

```python
def model(y,x):
    dydx=2*y+np.exp(2*x)
    return dydx
y0=3
x=np.linspace(0,1,100)
y=odeint(model,y0,x)

plt.style.use('seaborn')
plt.figure(figsize=(15,6))
plt.plot(x,y)
plt.xlabel("x")
plt.ylabel("y")
plt.show()
```
