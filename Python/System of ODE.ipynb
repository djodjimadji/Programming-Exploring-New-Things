Write python code to solve the following system of ordinary differential equations using the python function odeint.

```python
import numpy as np
from scipy.integrate import odeint, quad
import matplotlib.pyplot as plt
```

$$
\begin{cases}
\dfrac{dx}{dt}& = & e^t\\
\dfrac{dy}{dt}& = & x+2t\\
\dfrac{dz}{dt}& = & 3y-10t
\end{cases}
$$

```python
def fonction(u,t):
    x,y,z=u
    dxdt=np.exp(t)
    dydt=x+2*t
    dzdt=3*y-10*t
    return np.array([dxdt,dydt,dzdt])
# Initial condition
u0=np.array([1,-2,0])
a = 0
b = 1
N = 100
# time points
t = np.linspace(a,b,N+1)
# Solve ODE
u=odeint(fonction,u0,t)
x,y,z=u[:,0],u[:,1],u[:,2]
```

Exact solution

$$
\begin{cases}
x(t)& = & c_1e^t\\
y(t)& = & c_1e^t + t^2 +c_2\\
z(t)& = & 3c_1e^t+t^3+3c_2t-5t^2+c_3
\end{cases}
$$

Initial conditions

$$x(0) = 1, y(0) = -2, z(0) = 0$$

Particular solution

$$c_1 = 1, c_2 = -2, c_3 = -3$$

$$
\begin{cases}
x(t)& = & e^t\\
y(t)& = & e^t+t^2-3\\
z(t)& = & 3e^t+t^3-9t-5t^2-3
\end{cases}
$$

```python
x_exact=np.exp(t)
y_exact=np.exp(t)+t**2-3
z_exact=3*np.exp(t)+t**3-9*t-5*t**2-3
```

```python
plt.style.use('seaborn')
plt.figure(figsize=(15,6))

plt.subplot(2,2,1)
plt.plot(t,x)
plt.plot(t,x_exact,":.")
plt.xlabel("x")
plt.ylabel("t")
plt.legend(["Approximative solution","Exact solution"])


# plt.style.use('seaborn')
# plt.figure(figsize=(15,6))
plt.subplot(2,2,2)
plt.plot(t,y)
plt.plot(t,y_exact,":.")
plt.xlabel("t")
plt.ylabel("y")
plt.legend(["Approximative solution","Exact solution"])



# plt.style.use('seaborn')
# plt.figure(figsize=(15,6))
plt.subplot(2,2,3)
plt.plot(t,z)
plt.plot(t,z_exact,":.")
plt.xlabel("t")
plt.ylabel("z")
plt.legend(["Approximative solution","Exact solution"])
plt.show()
```
